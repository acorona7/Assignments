```python
# Removing null values and replacing with more descriptive text (object) or numbers (int or float)

NFL_df.time.fillna(0, inplace=True)
NFL_df.TimeSecs.fillna(0, inplace=True)
NFL_df.PlayTimeDiff.fillna(0, inplace=True)
NFL_df.SideofField.fillna('NoSide', inplace=True)
NFL_df.yrdln.fillna(0, inplace=True)
NFL_df.yrdline100.fillna(0, inplace=True)
NFL_df.GoalToGo.fillna(0, inplace=True)
NFL_df.posteam.fillna('NoPossession', inplace=True)
NFL_df.DefensiveTeam.fillna('NoPossession', inplace=True)
NFL_df.desc.fillna('NoDescription', inplace=True)
NFL_df.PosTeamScore.fillna('NoScore', inplace=True)
NFL_df.DefTeamScore.fillna('NoScore', inplace=True)
NFL_df.ScoreDiff.fillna('NoScoreDiff', inplace=True)
NFL_df.AbsScoreDiff.fillna('NoScoreDiff', inplace=True)


NFL_df.down.fillna(0, inplace=True)
NFL_df.FirstDown.fillna(0, inplace=True)
NFL_df.ExPointResult.fillna(0, inplace=True)
NFL_df.TwoPointConv.fillna(0, inplace=True)
NFL_df.DefTwoPoint.fillna(0, inplace=True)
NFL_df.PuntResult.fillna(0, inplace=True)
NFL_df.Passer.fillna('NoPasser', inplace=True)
NFL_df.DefTwoPoint.fillna(0, inplace=True)
NFL_df.PassOutcome.fillna('NoPass', inplace=True)
NFL_df.PassLength.fillna('NoPass', inplace=True)
NFL_df.PassLocation.fillna('NoPass', inplace=True)
NFL_df.Interceptor.fillna('NoPick', inplace=True)
NFL_df.Rusher.fillna('NoRush', inplace=True)
NFL_df.Rusher_ID.fillna('NoRush', inplace=True)
NFL_df.RunLocation.fillna('NoRush', inplace=True)
NFL_df.RunGap.fillna('NoRush', inplace=True)
NFL_df.Receiver.fillna('NoReceiver', inplace=True)
NFL_df.ReturnResult.fillna('NoReturn', inplace=True)
NFL_df.Returner.fillna('NoReturner', inplace=True)
NFL_df.BlockingPlayer.fillna('NoBlocker', inplace=True)
NFL_df.Tackler1.fillna('NoTackler', inplace=True)
NFL_df.Tackler2.fillna('No2ndTackler', inplace=True)
NFL_df.FieldGoalResult.fillna('NoFG', inplace=True)
NFL_df.FieldGoalDistance.fillna('NoFG', inplace=True)
NFL_df.RecFumbTeam.fillna('NoFumble', inplace=True)
NFL_df.RecFumbPlayer.fillna('NoFumble', inplace=True)
NFL_df.ChalReplayResult.fillna('NoChallenge', inplace=True)
NFL_df.PenalizedTeam.fillna('NoPenalty', inplace=True)
NFL_df.PenaltyType.fillna('NoPenalty', inplace=True)
NFL_df.PenalizedPlayer.fillna('NoPenalty', inplace=True)
NFL_df.DefTwoPoint.fillna(0, inplace=True)
NFL_df.airEPA.fillna(0, inplace=True)
NFL_df.yacEPA.fillna(0, inplace=True)
NFL_df.airWPA.fillna(0, inplace=True)
NFL_df.yacWPA.fillna(0, inplace=True)
```
